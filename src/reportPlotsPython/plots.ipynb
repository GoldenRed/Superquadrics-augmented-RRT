```python
from mpl_toolkits import mplot3d
%matplotlib inline 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

ETA = np.linspace(-np.pi/2, np.pi/2, 500)
OMEGA = np.linspace(-np.pi, np.pi, 500)
a = 1
b = 2
c = 0.5
eps1 = 1
eps2 = 1
X = []
Y = []
Z = []
for e in ETA:
    for o in OMEGA:
        X.append(a*(np.cos(e)**eps1)*(np.cos(o)**eps2))
        Y.append(b*(np.cos(e)**eps1)*(np.sin(o)**eps2))
        Z.append(c*(np.sin(e)**eps1))

```

```python
fig = plt.figure(figsize=(40,40))
ax = Axes3D(fig)
ax.set_xlim3d(-2,2)
ax.set_ylim3d(-2,2)
ax.set_zlim3d(-2,2)


ax.scatter(X,Y,Z)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

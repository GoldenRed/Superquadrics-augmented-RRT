```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
def fexp(base, power):
    """the signed exponential function"""
    return (np.sign(base) * (np.abs(base)**power))
# Data for plotting
t = np.arange(-np.pi, np.pi, 0.1)

```

```python
eps = 0.2
x = 1*fexp(np.cos(t),eps)
y = 1*fexp(np.sin(t),eps)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(x, y, 'ko')

ax.set(xlabel=r"x-value", ylabel=r"y-value",
       title=r"Superellipse, $\epsilon=0.2$")
ax.grid()


plt.axis('equal')
fig.savefig("Se_Squareish.pdf", format='pdf')
plt.show()
```
